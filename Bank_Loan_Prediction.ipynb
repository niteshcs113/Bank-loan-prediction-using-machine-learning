```python
import numpy as np
import pandas as pd

data=pd.read_excel("E://Loan_Data//loan.xlsx")
```

```python
data.shape,data.columns,data.dtypes
```

```python
data['term'].replace(to_replace=' months', value='', regex=True, inplace=True)
data['term'] = pd.to_numeric(data['term'])
```

```python
data.head()
```

```python
cols=data.columns
for i in cols:
    if data[i].dtype==np.object:
        s=set(data[i])
        d={}
        for i1,j in enumerate(s):
            d[j]=i1
        data[i]=list(map(lambda k:d[k],data[i]))
```

```python
data.head()
```

```python
data.fillna(0,inplace=True)
```

```python
corr=pd.DataFrame(data.corr())
corr.head(1)
```

```python
dt=[]
for i,j in enumerate(corr.columns):
    if corr[j][0]>0 and corr[j][0]!=1 and corr[j][0]!=np.nan:
        dt.append(corr.columns[i])
print(dt)
```

```python
dt=data[['member_id', 'loan_amnt', 'funded_amnt', 'funded_amnt_inv', 'term', 'int_rate', 'installment', 'sub_grade', 'home_ownership', 'annual_inc', 'loan_status', 'url', 'purpose', 'addr_state', 'dti', 'open_acc', 'revol_bal', 'revol_util', 'total_acc', 'out_prncp', 'out_prncp_inv', 'total_pymnt', 'total_pymnt_inv', 'total_rec_prncp', 'total_rec_int', 'recoveries', 'last_pymnt_amnt']]
```

```python
dt.shape
```

```python
x=dt.drop(['loan_status'],1)
y=dt['loan_status']
```

```python
x.shape,y.shape
```

```python
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVR
x=scale(x)
x_tr,x_ts,y_tr,y_ts=train_test_split(x,y,train_size=0.75)
algo=LogisticRegression()
algo.fit(x_tr,y_tr)
print("Accuracy is",algo.score(x_ts,y_ts))
```

```python
from sklearn.svm import SVC
alg=SVC()
alg.fit(x_tr,y_tr)
print("Accuracy by SVC is",alg.score(x_ts,y_ts))
```

```python
from sklearn.neighbors import KNeighborsClassifier
alg1=KNeighborsClassifier()
alg1.fit(x_tr,y_tr)
print("Accuracy by KNN is",alg1.score(x_ts,y_ts))
```

```python
from sklearn.naive_bayes import GaussianNB
alg2=GaussianNB()
alg2.fit(x_tr,y_tr)
print("Accuracy by Naive Bayes is",alg2.score(x_ts,y_ts))
```

```python
from sklearn.tree import DecisionTreeClassifier
alg3=DecisionTreeClassifier()
alg3.fit(x_tr,y_tr)
print("Accuracy by Decision Tree is",alg3.score(x_ts,y_ts))
```

```python
from sklearn.ensemble import RandomForestClassifier
alg4=RandomForestClassifier()
alg4.fit(x_tr,y_tr)
print("Accuracy by Random Forest is",alg4.score(x_ts,y_ts))
```

```python
import xgboost
alg1 = xgboost.sklearn.XGBClassifier(
    objective="binary:logistic", 
    learning_rate=0.05, 
    seed=9616, 
    max_depth=20, 
    gamma=10, 
    n_estimators=500)

alg1.fit(x_tr,y_tr)
print("Acurracy is ",alg1.score(x_ts,y_ts))
```
